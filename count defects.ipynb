## Program to count amount of defects on images from predictions generated by Predictions.ipynb

```python
import csv
# Open file to read from given path
with open('Downloads/test1.csv', 'rt') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    
    defect_labels = []
    
    for row in reader:
         defect_labels.append(row)
defect_labels
```

```python
def count_defects (defect_labels):
    defect_counts = []
    
    ## With different images modifications for next 3 lines are required
    image_number = 1028
    last_image_number = 1100
    image_name = 'Acrorad_0712-1001-1-07-0' + str(image_number) + '.jpg'
    
    row_number = 1
    count = 0
    
    while image_number <= last_image_number:        
        if row_number == len(defect_labels):
            defect_counts.append([image_name, 0])
        elif not defect_labels[row_number][1] == image_name:
            defect_counts.append([image_name, 0])
        else:
            while defect_labels[row_number][1] == image_name:
                count += 1
                row_number += 1
                ## to avoid IndexError
                if row_number == len(defect_labels):
                    break
            defect_counts.append([image_name, count])
            count = 0
            
        image_number += 1
        image_name = 'Acrorad_0712-1001-1-07-0' + str(image_number) + '.jpg'    
            
    return defect_counts
```

```python
defect_counts = count_defects(defect_labels)
```

```python
import csv
with open('Acrorad_0712-1001-1-07-0(1028-1100)-defect_counts.csv', 'wt') as csvfile:
    writer = csv.writer(csvfile)
    
    #Write headers
    writer.writerow(['image', 'amount of defects'])
    
    #Write content
    for row in defect_counts:
        writer.writerow(row)
```
